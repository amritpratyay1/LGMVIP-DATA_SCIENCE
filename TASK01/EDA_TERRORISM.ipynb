## EDA

### CLEARING DATA 

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                       'region_txt':'Region','city':'City','latitude':'Latitude','longitude':'Longitude','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
data=data[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
data.head()
```


### HANDLING MISSING DATA

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap="v")
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
data['Wounded'] = data['Wounded'].fillna(0)
data['Killed'] = data['Killed'].fillna(0)
data['Casualities'] = data['Killed'] + data['Wounded']
```

```python
data['State'].fillna("unknown", inplace = True)
data['City'].fillna("unknown", inplace = True)
```

```python
data.dropna(axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
print("Most attack month:",data['Month'].value_counts().idxmax())
print("Most attack Year:",data['Year'].value_counts().idxmax())
print("Most attack City:",data['City'].value_counts().index[1])
print("Most attack Country",data['Country'].value_counts().idxmax())
print("Region with the most attacks:",data['Region'].value_counts().idxmax())
print("Group with the most attacks:",data['Group'].value_counts().index[1])
print("Most Attack Types:",data['AttackType'].value_counts().idxmax())
```

```python
data['Year'].value_counts(dropna = False).sort_index()
```

### DATA VISUALISATION

```python
plt.figure(figsize=[10,8])
sns.heatmap(data.corr(),cmap='Blues', linewidths=0.4,annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = 15,11
sns.barplot(x = data["Country"].value_counts()[:20].index, y=data["Country"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", size=20)
plt.xticks(rotation=90)
plt.xlabel("Country", size=20)
plt.title("COUNTRY WISE ATTACKS", size=25, fontweight="bold")
```

```python
top_countries = data.Country.value_counts().head(20)
top_countries
```

```python
plt.pie(top_countries, labels=top_countries.index)
plt.title("Top 20 Countries with most terror attack")
plt.show()
```

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.ylabel('Numbers of Attacks',size=20)
plt.xlabel("Year of attack",size=20)
plt.title('TERRORIST ACTIVITIES BY REGION',size=25,fontweight="bold")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = 20,10
sns.barplot(x = data["Month"].value_counts()[:20].index, y=data["Month"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", size=18)
plt.xticks(rotation=90)
plt.xlabel("Month of Attack", size=20)
plt.title(" ATTACKS BY MONTH ", size=25, fontweight="bold")
plt.show()
```

```python
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,y = years_count,palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Year',size=20)
plt.ylabel('Number of Attacks',size=20)
plt.title('NUMBER OF ATTACKS YEAR WISE',size=25,fontweight="bold")
plt.show()
```

```python
df = data[['Year','Killed']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.ylabel("Number of people killed",size=20)
plt.xlabel('Year',fontsize=20)
plt.title("PEOPLE DIED IN ATTACK",size=25,fontweight="bold")
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
plt.rcParams["figure.figsize"] = 20, 8
sns.barplot(x = data["Group"].value_counts()[:20].index, y=data["Group"].value_counts()[:20].values)
plt.ylabel("Total Attack", size=20)
plt.xticks(rotation=90)
plt.xlabel("Attack Group", size=20)
plt.title("ATTACKS BY GROUP", size=25, fontweight="bold")
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color=(0.2, 0.7, 0.3, 0.8))
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.title("TOP 10 MOST EFFECTED CITIES",size=25,fontweight="bold")
plt.show()
```

```python
data[['AttackType','Killed']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,8),color=(0.1, 0.5, 0.4, 0.9))
plt.xticks(rotation=50)
plt.ylabel('Number of people killed',fontsize=20)
plt.xlabel('Attack type',fontsize=20)
plt.title("NUMBER OF PEOPLE KILLED IN DIFFERENT ATTACK TYPES",fontsize=25,fontweight="bold")
plt.show()
```
